# Bag of words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["people watch somesh",
                         "somesh watch somesh",
                         "people write comment",
                          "somesh write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#vocabulary
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
# new
cv.transform(['Somesh watch Trivedi']).toarray()
```

#  N-grams

```python
df = pd.DataFrame({"text":["people watch somesh",
                         "somesh watch somesh",
                         "people write comment",
                          "somesh write comment"],"output":[1,1,0,0]})


df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

# TF-IDF (Term frequency- Inverse document frequency)

```python
df = pd.DataFrame({"text":["people watch somesh",
                         "somesh watch somesh",
                         "people write comment",
                          "somesh write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid= TfidfVectorizer()
```

```python
tfid.fit_transform(df['text']).toarray()
```

```python
print(tfid.idf_)
```

```python
print(tfid.get_feature_names_out())
```
